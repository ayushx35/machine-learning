```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from  sklearn import linear_model
```

```python
fl=pd.read_csv('data/canada_per_capita_income.csv')
fl
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(fl.year,fl['per capita income (US$)'],color='red',marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(fl[['year']],fl['per capita income (US$)'])
```

```python
reg.predict(pd.DataFrame([[5000]], columns=['year']))
```
